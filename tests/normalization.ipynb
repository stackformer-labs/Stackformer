```python
import torch
import torch.nn as nn
import torch.nn.functional as F
```

Norm = (Xn - Mean) / sqrt(var) + Bias

output= γ x Norn + β

```python
emb_dim = 5
x = torch.randn(emb_dim,emb_dim)
print("Input: \n",x)
mean = x.mean(dim=-1, keepdim=True)
print("mean: \n", mean)
var = x.var(dim=-1, keepdim=True, unbiased=False)
print("Var: \n",var)
norm_x = (x - mean) / torch.sqrt(var)
print("Normalized: \n", norm_x)
scale = nn.Parameter(torch.ones(emb_dim))
shift = nn.Parameter(torch.zeros(emb_dim))
print("output: \n",scale * norm_x + shift)
```

```python
class LayerNorm(nn.Module):
    def __init__(self, emb_dim, eps = 1e-5):
        super().__init__()
        self.eps = eps
        
    def forward(self, x):
        mean = x.mean(dim=-1, keepdim=True)
        var = x.var(dim=-1, keepdim=True, unbiased=False)
        norm_x = (x - mean) / torch.sqrt(var + self.eps)
        return norm_x 
    
x = torch.randn(2,10,64)
layer_noem = LayerNorm(64)
out = layer_noem(x)
print("LayerNormilization:", out.shape)
```

2) RMS Normalization \
RMS = sqrt(Xn ** 2) \
Norm = Xn / RMS

```python
emb_dim = 5
x = torch.randn(emb_dim,emb_dim)
print("input:\n",x)
norm = x.pow(2).mean(-1,keepdim=True)
print("norm\n",norm)
rms = norm.sqrt()
print(x / rms)
```

```python
class RMSNormilization(nn.Module):
    
    def __init__(self,dim,eps=1e-8):
        super().__init__()
        self.eps = eps
        self.weight = nn.Parameter(torch.ones(dim))
        
    def forward(self,x):
        rms = x.pow(2).mean(-1,keepdim=True).sqrt()
        return self.weight * x / (rms + self.eps)
    
x = torch.randn(2,10,64)
rms = RMSNormilization(64)
out = rms(x)
print("RMSNormilization:", out.shape)
```
