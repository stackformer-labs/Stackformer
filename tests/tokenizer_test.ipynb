```python
# letter tokenizer
def letter_tokenizer(data):
    vocab = sorted(set(data))  # sorted list of unique characters
    letter2idx = {letter: idx for idx, letter in enumerate(vocab)}  # mapping
    return [letter2idx[letter] for letter in data]
```

```python
# Bite pair (BPE) Embedding using tokienizer
def Embedding_using_tiktoken(data,vocab_size,embedding_dim,model: str):
    # 'gpt2', 'r50k_base', 'p50k_base', 'p50k_edit', 'cl100k_base', 'o200k_base'
    Embedding = tiktoken.get_encoding(model)
    tokens = Embedding.encode(data)
    embedding_layer = nn.Embedding(num_embeddings=vocab_size, embedding_dim=embedding_dim)
    tensors = torch.tensor(tokens)
    embedded = embedding_layer(tensors)
    return embedded
```
